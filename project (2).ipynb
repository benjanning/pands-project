# Machine Learning

```python
# import load_iris function from datasets module
from sklearn.datasets import load_iris
```

```python
# save "bunch" object (Scikit special container object exposing keys as attributes)
# containing iris dataset and all its attributes:
iris = load_iris()
type(iris)
```

```python
# print iris data
print(iris.data)
```

# 2 Types of Machine Learning Problems

```python
# print the names of the four features:
iris.feature_names
```

```python
# print integers representing the species of observation:
# we will refer to them as responses.

# We need to convert to numbers as sckitlearn will not accept strings.

iris.target
```

```python
# print the encoding scheme for species: 
#   0 = setosa
#   1 = versicolor
#   2 = virginica
iris.target_names
```

# Requirements for working with dat in scikit-learn:

```python
# check the types of feature and response:
print(type(iris.data))
print(type(iris.target))
```

```python
# Check shape of features (first dimension = number of observations, second dimension = number of features)
print(iris.data.shape)
```

```python
# Check the shape of the response (single dimension matching the number of observations)
print(iris.target.shape)
```

```python
# SKLearn convention requires feaature and response to be stored as the following:

# Notice X is uppercase (as it represents a matrix), while y is lowercase.

#store feature matrix in "x"
X = iris.data

# Store response vector in "y"
y = iris.target
```

# K-nearest Neighbors (KNN) CLassification

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Create classifier object that looks for the 1 neares neighbor
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X, y)
```

```python
# Important note: Needs 2 sets of square brackets.
knn.predict([[3, 5, 4, 2]])
```

```python
knn5 = KNeighborsClassifier(n_neighbors=5)
knn5.fit(X, y)
knn5.predict([[3, 5, 4, 2]])
```
